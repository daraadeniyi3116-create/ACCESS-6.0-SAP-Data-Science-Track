## PART B - Univariate Analysis


#### Numerical Analysis

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['RETAIL SALES'], bins=50, kde=False, color='blue')
plt.title('Distribution of RETAIL SALES')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['RETAIL SALES'], color='orange')
plt.title('Boxplot of RETAIL SALES')
plt.show()
```

1. **What does the distribution look like?** It looks heavily skewed to the right. The histogram has one huge blue bar on the left (meaning almost all sales are very small amounts near 0), and then a tiny tail that stretches out to the right. Its definitely not a normal bell shape.

2. **Are there possible outliers?** Yes, definitely. The boxplot shows a thick line of dots stretching way out to the right, going past 2,500. Since most of the sales are really small, all those dots represent sales that are unusually high compared to the rest.

#### Categorical Analysis

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='YEAR', data=df, palette='viridis')
plt.title('Frequency of Transactions by YEAR')
plt.show()
```

1. **Which category appears most often?** The year 2019 appears the most often by far. It has over 130,000 transactions, which is way more than any other year.

2. **What could this mean?** It looks like 2019 was their peak year for activity. The big drop in 2020 might be because of the pandemic or maybe the data stops halfway through that year.

#### PART C - Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='YEAR', y='RETAIL SALES', data=df, showfliers=False)
plt.title('Retail Sales by Year')
plt.xlabel('Year')
plt.ylabel('Retail Sales ($)')
plt.show()
```

```python
print("Summary Stats for Warehouse Sales:")
print(df['WAREHOUSE SALES'].describe())


plt.figure(figsize=(8, 4))
sns.boxplot(x=df['WAREHOUSE SALES'], color='red')
plt.title('Outliers in WAREHOUSE SALES')
plt.show()
```

1. **Why might these outliers exist?** The boxplot shows a crazy amount of dots stretching way out to the right, going all the way up to 18,000. Since the average sale is only around 25, these huge numbers are definitely outliers. They exist because this is a warehouse, so while most orders are small, occasionally a big business comes in and buys a massive amount of stock at once.

2. **Are they likely errors or genuine business events?**

The High Numbers (Genuine): The values going up to 18,000 are likely genuine business events. It makes sense for a warehouse to have some giant bulk orders.

The Low Numbers (Errors): However, there are dots on the left side going down to -7,800. You can't really have "negative sales" unless it's a return, but a return of -7,800 seems super suspicious. That one is probably a data entry error.

## Part E - Correlation Analysis.

```python
numeric_data = df[['RETAIL SALES', 'WAREHOUSE SALES']]

corr_matrix = numeric_data.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation: Retail vs Warehouse')
plt.show()
```

**1. Is the correlation weak, moderate, or strong?** The correlation is 0.49, which is moderate. Its basically right in the middlenot super strong (it's not near 1.0) but not totally weak (it's not 0). It shows there is some connection, but it's not a perfect match.

**2. How could this information be useful in business analysis?** This is important because it tells us that Retail Sales and Warehouse Sales don't always happen at the exact same time. Since the correlation isn't perfect, the business can't just look at shop sales to know exactly what the warehouse is doing. They need to manage the warehouse inventory separately because bulk orders might happen even when retail shops are quiet.

##PART F - Business Questions & Insights

```python

# Q1: Which month has the highest Retail Sales? (Seasonality)
# grouping by month to see when people buy the most
monthly_sales = df.groupby('MONTH')['RETAIL SALES'].sum()

plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Total Retail Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Q2: Are we selling more or less over the years? (Trend)
yearly_sales = df.groupby('YEAR')['RETAIL SALES'].sum()

plt.figure(figsize=(8, 5))
sns.barplot(x=yearly_sales.index, y=yearly_sales.values, palette='Blues')
plt.title('Total Retail Sales per Year')
plt.show()

# Q3: Where does most of our money come from? (Retail vs Warehouse)
total_retail = df['RETAIL SALES'].sum()
total_warehouse = df['WAREHOUSE SALES'].sum()

plt.figure(figsize=(6, 6))
plt.pie([total_retail, total_warehouse], labels=['Retail', 'Warehouse'], autopct='%1.1f%%', colors=['#ff9999', '#66b3ff'])
plt.title('Revenue Split: Retail vs Warehouse')
plt.show()
```

**1. Which month is busiest?**

Looking at the line chart, sales definitely peek in specific months (likely summer or holiday seasons).

Insight: We should make sure the warehouse is fully stocked a month before these peaks so we don't run out.

**2. Is the business growing?**

The bar chart shows how total sales change each year.

Insight: If the bars are getting taller, we are doing great. If they are shrinking, we might need more marketing.

**3. Retail vs Warehouse?**

The pie chart shows the split.

Insight: If the Warehouse slice is huge, we are basically a wholesale company. If Retail is bigger, we are a consumer brand. We should focus our budget on the bigger slice.
