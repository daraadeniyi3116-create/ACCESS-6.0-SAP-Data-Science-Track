```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Warehouse_and_Retail_Sales.csv')

df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.fillna(0)
```

```python
df['TOTAL_SALES'] = df['RETAIL SALES'] + df['WAREHOUSE SALES']
```

```python
df['RETAIL_RATIO'] = df['RETAIL SALES'] / (df['TOTAL_SALES'] + 0.001)
```

```python
df[['RETAIL SALES', 'WAREHOUSE SALES', 'TOTAL_SALES', 'RETAIL_RATIO']].head()
```

```python
df.to_csv('Cleaned_Sales_Data.csv', index=False)

print("File saved! Check the folder icon on the left to download it.")
```
