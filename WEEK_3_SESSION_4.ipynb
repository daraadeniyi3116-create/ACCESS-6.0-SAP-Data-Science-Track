## Handling Missing Values


# Missing Values
1. **SUPPLIER**: 167 missing.

SOLUTION; Fill with "Unknown", since 167 is less that 0.05% of the data.

2. **RETAIL SALES**: 3 missing.

SOLUTION; Fill with Zero (0), since its better than remove the whole entry.

3. **ITEM TYPE**: 1 missing.

SOLUTION; Remove the entire entry, it is irrelevant.

```python
df['SUPPLIER'] = df['SUPPLIER'].fillna('Unknown')

df = df.dropna(subset=['ITEM TYPE'])

df['RETAIL SALES'] = df['RETAIL SALES'].fillna(0.0)
```

```python
print("Remaining missing values:", df.isnull().sum().sum())
```

There are no Duplicates


```python
df['TOTAL_SALES'] = df['RETAIL SALES'] + df['WAREHOUSE SALES']

df['RETAIL_SHARE'] = np.where(df['TOTAL_SALES'] == 0, 0, df['RETAIL SALES'] / df['TOTAL_SALES'])
```

```python
df['NON_REVENUE_MOVE'] = df['RETAIL TRANSFERS'].abs()
```
